# EXTRACT 

## Read CSV

```python
import pandas as pd
doctors=pd.read_csv("../Raw Datas/doctors.csv")
patients = pd.read_csv("../Raw Datas/patients.csv")
appointments = pd.read_csv("../Raw Datas/appointments.csv")
treatments = pd.read_csv("../Raw Datas/treatments.csv")
billing = pd.read_csv("../Raw Datas/billing.csv")

patients.head()
```

# Transform

## Checking data types

```python
doctors.dtypes
```

## Changing date type of patients

```python
if(patients["date_of_birth"].dtype != "datetime64[ns]"):
    patients["date_of_birth"] = pd.to_datetime(patients["date_of_birth"]);

if(patients["registration_date"].dtype != "datetime64[ns]"):
    patients["registration_date"] = pd.to_datetime(patients["registration_date"]);
patients.dtypes
```

## Changing big integer values

```python
if(patients["contact_number"].dtype != "object"):
    patients["contact_number"] = patients["contact_number"].astype("object");
if(doctors["phone_number"].dtype != "object"):
    doctors["phone_number"] = doctors["phone_number"].astype("object");
doctors.dtypes
```

## Changing date type of appointment

```python
if(appointments["appointment_date"].dtype != "datetime64[ns]"):
    appointments["appointment_date"] = pd.to_datetime(appointments["appointment_date"]);

appointments["appointment_time"] = pd.to_datetime(
    appointments["appointment_time"],
    format="%H:%M:%S",
    errors="coerce"
).dt.time

print(type(appointments["appointment_time"].iloc[0]))
```

## Changing date type of Treatment

```python
if(treatments["treatment_date"].dtype != "datetype64[ns]"):
    treatments["treatment_date"] = pd.to_datetime(treatments["treatment_date"]);
treatments.dtypes
```

## Changing date type of Billing

```python
if(billing["bill_date"].dtype != "datetime64[ns]"):
    billing["bill_date"] = pd.to_datetime(billing["bill_date"]);
billing.dtypes
```

## Removing column patient_id from Billing as normalization

```python
billing = billing.drop(columns=["patient_id"]);
billing.columns
```

## Checking null values

```python
print(patients.isnull().sum());
print(doctors.isnull().sum());
print(appointments.isnull().sum());
print(treatments.isnull().sum());
print(billing.isnull().sum());
```

## Checking duplicates

```python
print(patients.duplicated().sum());
print(doctors.duplicated().sum());
print(appointments.duplicated().sum());
print(treatments.duplicated().sum());
print(billing.duplicated().sum());
```

## Checking Outliers

```python
print(doctors["years_experience"].describe()); 
print(treatments["cost"].describe());
print(billing["amount"].describe());
```

## Lowercasing each column in case it doesn't match SQL columns

```python
patients.columns = patients.columns.str.lower() ;
doctors.columns = doctors.columns.str.lower() ;
appointments.columns = appointments.columns.str.lower() ;
treatments.columns = treatments.columns.str.lower() ;
billing.columns = billing.columns.str.lower() ;
```

# Load

```python
patients.to_sql("Patients" , engine, if_exists="append", index=False);
doctors.to_sql("Doctor", engine, if_exists="append", index=False);
appointments.to_sql("Appointments", engine, if_exists="append" , index=False);
treatments.to_sql("Treatments", engine, if_exists="append", index=False);
billing.to_sql("Billing", engine, if_exists="append", index=False);
```
